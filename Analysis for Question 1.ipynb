<p> Examining the summary anlysis for order_amount and total_item, it appears the error stemmed from calculating the average order value by taking the average of the order_amount column (3145.13).</p>


<p> According to this <a optimizely.com/optimization-glossary/average-order-value/#:~:text=To%20calculate%20your%20company's%20average,September's%20monthly%20AOV%20was%20%2431" target="_blank">optimizely.com</a>, Average order value (AOV) = the total revenue divided by the total number of orders, so Calculating the average order value by finding the average order_amount is wrong.</p> 


<p> I created a dataframe 'aov' to hold the average order value of each 100 stores </p>

```python
aov = df.groupby(['shop_id'])['order_amount', 'total_items'].sum()
```

```python
aov['average_order_value'] = aov['order_amount'] / aov['total_items']
aov
```

<p> Using summary staticts on the "average_order_value" column </P>

```python
aov['average_order_value'].describe()
```

<p> AOV of $407.99 looks realistic but we know shopify sells affordable shoes, so the average order value is still high. <br>
There is also a huge gap between the minimum and the maximum order value.

```python
pd.set_option("max_rows", None)
aov.sort_values(by='average_order_value', ascending=False)
```

<p> Sorting the dataframe by average_order_value column in desecinding order reveals that shop id 78 only sold 88 pairs of sneakers to make 2263800 dollars. <br> <br>
Since each of these shops sells only one model of shoe, we know average order value = cost of shoe. <br> 
store id 78's selling sneakers that cost 25725.0 is quite unusual, it is an outlier. <br> <br>
We can use the mode and the medain function to reduce the effect of store 78 on the analysis </p>


```python
# checking for median
aov['average_order_value'].median()
```

```python
# checking for mode
aov['average_order_value'].mode()
```

<p> The median of the average order value of the 100 shops is 153 and the mode of the the average order value of the 100 shops is also 153 <br> <br>
I would report the mode of the average order value it reduces the effect of outliers. <br> <br>
